```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
file_path = "/home/yichi/research/ad_afqmc/ad_afqmc/corr_sample_test/ammonia/w600r29s29.data"
en_diff_mean = []
en_diff_mean_idx = 5
en_diff_mean_err = []
en_diff_mean_err_idx = 6


with open(file_path, "r") as file:
    for line in file:
        if line.strip() and not line.startswith("#"): 
            columns = line.split()  
            if len(columns) > en_diff_mean_idx:
                en_diff_mean.append(columns[en_diff_mean_idx])
            if len(columns) > en_diff_mean_err_idx:
                en_diff_mean_err.append(columns[en_diff_mean_err_idx])

en_diff_mean = np.array(en_diff_mean,dtype='float32')
print(en_diff_mean)
# for i,elmt in enumerate(en_diff_mean_err):
#     if elmt == 'None':
#         en_diff_mean_err[i] = '0'
en_diff_mean_err = np.array(en_diff_mean_err,dtype='float32')
print(en_diff_mean_err)

tot_walker = 600
n_run = 200

prop_steps = np.arange(30)

afqmc_en1 = -112.8194
afqmc_en1_err = 0.0002
afqmc_en2 = -112.8119
afqmc_en2_err = 0.0003
afqmc_en_diff = afqmc_en1-afqmc_en2
afqmc_en_diff_err = np.sqrt(afqmc_en1_err**2+afqmc_en2_err**2)
x_steps = np.linspace(0,max(prop_steps),100)
afqmc_en = np.array([afqmc_en_diff]*len(x_steps))
afqmc_lw = afqmc_en-afqmc_en_diff_err
afqmc_up = afqmc_en+afqmc_en_diff_err
plt.plot(x_steps,afqmc_en)
plt.fill_between(x_steps, afqmc_lw, afqmc_up, color='blue', alpha=0.3, label='afqmc')

plt.errorbar(prop_steps,en_diff_mean,yerr=en_diff_mean_err, fmt='o', capsize=5,label='cs')
plt.xlabel('steps')
plt.ylabel('energy difference')
plt.title('correlated walkers energy differences')
plt.legend()
plt.show()
```
