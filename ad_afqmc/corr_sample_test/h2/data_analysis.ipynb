```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
n_walker_pp = 30
nproc = 10
n_run = 200

eql_steps = [1,2,3,5,7,11,13]
en_diff = [-0.009233,-0.008141,-0.007566,-0.005911,-0.005553,-0.005983,-0.005610]
error = [0.000134,0.000352,0.005971,0.001165,0.001785,0.001684,0.002394]

afqmc_en_diff = -0.005606581337333716
afqmc_en_diff_err = 0.0019502384765517759
x_steps = np.linspace(1,13,100)
afqmc_en = np.array([afqmc_en_diff]*len(x_steps))
afqmc_lw = afqmc_en-afqmc_en_diff_err
afqmc_up = afqmc_en+afqmc_en_diff_err
plt.plot(x_steps,afqmc_en)
plt.fill_between(x_steps, afqmc_lw, afqmc_up, color='blue', alpha=0.3, label='afqmc energy')

plt.errorbar(eql_steps,en_diff,yerr=error, fmt='o', capsize=5,label='cs')
plt.xlabel('steps')
plt.ylabel('energy difference')
plt.title('correlated walkers energy differences')
plt.legend()
plt.show()
```

```python
file_path = "w_40_r_200_s_29.out"
en_diff_mean = []
en_diff_mean_idx = 5
en_diff_mean_err = []
en_diff_mean_err_idx = 6


with open(file_path, "r") as file:
    for line in file:
        if line.strip() and not line.startswith("#"): 
            columns = line.split()  
            if len(columns) > en_diff_mean_idx:
                en_diff_mean.append(columns[en_diff_mean_idx])
            if len(columns) > en_diff_mean_err_idx:
                en_diff_mean_err.append(columns[en_diff_mean_err_idx])

en_diff_mean = np.array(en_diff_mean[2:],dtype='float32')
print(en_diff_mean)
for i,elmt in enumerate(en_diff_mean_err):
    if elmt == 'None':
        en_diff_mean_err[i] = '0'
en_diff_mean_err = np.array(en_diff_mean_err[2:],dtype='float32')
print(en_diff_mean_err)

n_walker_pp = 40
nproc = 10
n_run = 200

prop_steps = np.arange(30)

afqmc_en_diff = -0.005606581337333716
afqmc_en_diff_err = 0.0019502384765517759
x_steps = np.linspace(1,max(prop_steps),100)
afqmc_en = np.array([afqmc_en_diff]*len(x_steps))
afqmc_lw = afqmc_en-afqmc_en_diff_err
afqmc_up = afqmc_en+afqmc_en_diff_err
plt.plot(x_steps,afqmc_en)
plt.fill_between(x_steps, afqmc_lw, afqmc_up, color='blue', alpha=0.3, label='afqmc')

plt.errorbar(prop_steps,en_diff_mean,yerr=en_diff_mean_err, fmt='o', capsize=5,label='cs')
plt.xlabel('steps')
plt.ylabel('energy difference')
plt.title('correlated walkers energy differences')
plt.legend()
plt.show()
```
