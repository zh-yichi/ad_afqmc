```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# H2
def get_data(data):
    n = []
    e = []
    err = []
    for line in data.splitlines():
        if len(line) > 0:
            m = [int(line.split()[0])]
            n = np.hstack((n,m))
            energy = [float(line.split()[-3])]
            e = np.hstack((e,energy))
            error = [float(line.split()[-1])]
            err = np.hstack((err,error))
    e_per = e/n
    err_per = err/n
    return n,e_per,err_per
data1 = '''
1  AFQMC energy: -1.096072 +/- 0.00000
2  AFQMC energy: -2.19206 +/- 0.00005
4  AFQMC energy: -4.3838 +/- 0.0003
8  AFQMC energy: -8.7665 +/- 0.0003
16 AFQMC energy: -17.5296 +/- 0.0008
32 AFQMC energy: -35.051 +/- 0.002
50 AFQMC energy: -54.758 +/- 0.004
'''
data2 = '''
1  AFQMC/CCSD_PT energy: -1.096170 +/- 0.000017
2  AFQMC/CCSD_PT energy: -2.192323 +/- 0.000049
4  AFQMC/CCSD_PT energy: -4.384510 +/- 0.000177
8  AFQMC/CCSD_PT energy: -8.768645 +/- 0.000308
16 AFQMC/CCSD_PT energy: -17.538406 +/- 0.001528
32 AFQMC/CCSD_PT energy: -35.076499 +/- 0.001928
50 AFQMC/CCSD_PT energy: -54.812214 +/- 0.004169
'''
n1,e_per1,err_per1 = get_data(data1)
n2,e_per2,err_per2 = get_data(data2)
plt.errorbar(n1,e_per1,yerr=err_per1,fmt='.',capsize=5,label='AFQMC/CISD')
plt.plot(n1,e_per1, '--', color='C0', alpha=0.3)
plt.errorbar(n2,e_per2,yerr=err_per2,fmt='.',capsize=5,label='AFQMC/CCSD_PT')
plt.plot(n2,e_per2, '--', color='C1', alpha=0.3)
plt.xlabel('number of H2')
plt.ylabel('energy per H2')
plt.legend()
plt.show()
```
