```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
## water ##
# converged SCF energy = -76.0267469569854
# AFQMC energy: -76.24173 +/- 0.00006
```

```python
ccsd_afqmc_cr = -76.24173 + 76.0267469569854
print(ccsd_afqmc_cr)
ccsd_afqmc_err = 0.00006*2
```

```python
lno_thresh = np.array([0.0001,5e-05,3e-05,1e-05,1e-06])
lno_ccsd_afqmc_cr = np.array([-0.202613,-0.211251,-0.213498,-0.214472,-0.214940])-ccsd_afqmc_cr
lno_ccsd_afqmc_mp2_cr = np.array([-0.213817,-0.214611,-0.214893,-0.214856,-0.214945])-ccsd_afqmc_cr
lno_ccsd_afqmc_err = np.array([0.000043,0.000047,0.000046,0.000050,0.000052])*2
```

```python
x = np.linspace(max(lno_thresh),min(lno_thresh),100)
plt.plot(x,np.array([0]*100),color='black', linestyle='--')
plt.errorbar(lno_thresh,lno_ccsd_afqmc_cr,yerr=lno_ccsd_afqmc_err,fmt='o',capsize=5,label='lno_afqmc/ccsd')
plt.errorbar(lno_thresh,lno_ccsd_afqmc_mp2_cr,yerr=lno_ccsd_afqmc_err,fmt='o',capsize=5,label='lno_afqmc/ccsd_mp2')
plt.fill_between(x, np.array([0]*100) + ccsd_afqmc_err, np.array([0]*100) - ccsd_afqmc_err, color='blue', alpha=0.4,label='afqmc_cr')
plt.xlabel('lno_thresh')
plt.ylabel('energy (Hartree)')
plt.title('lno-afqmc/ccsd energy - afmqc/ccsd erngy')
plt.legend()
plt.show()
```
